```python

class Permiso:
    def __init__(self, nombre, nivel, descripcion):
        self.nombre = nombre
        self.nivel = nivel
        self.descripcion = descripcion

    def leer_datos(self):
        print("--- Ingreso de Información ---")
        self.nombre = input("Ingrese el nombre del permiso: ")
        self.nivel = input("Ingrese el nivel (Ej: Alto, Medio, Bajo): ")
        self.descripcion = input("Ingrese la descripción: ")

    def mostrar_datos(self):
        print("\n--- Datos del Permiso ---")
        print("Nombre      :", self.nombre)
        print("Nivel       :", self.nivel)
        print("Descripción :", self.descripcion)

# --- Creación de los objetos permiso1 y permiso2 ---

print("=======================================")
print("          REGISTRO DE PERMISOS         ")
print("=======================================")

# Objeto 1
print("\n--- Permiso Número 1 ---")
permiso1 = Permiso("", "", "") # Creando el objeto permiso1
permiso1.leer_datos()
permiso1.mostrar_datos()

print("---------------------------------------")

# Objeto 2
print("\n--- Permiso Número 2 ---")
permiso2 = Permiso("", "", "") # Creando el objeto permiso2
permiso2.leer_datos()
permiso2.mostrar_datos()

print("=======================================")
print("Software desarrollado Por Pedro Martinez NL = 34")
```
